```python
#Lowercasing text
df['content']=df['content'][3].lower()
```

```python
df['title']=df['title'].str.lower()
```

```python
df
```

```python
#removing tag from the text
import re
def remove_html_tags(df):
    pattern=re.compile('<.*?>')
    return pattern.sub(r'',df)
```

```python
df['content']=df['content'].apply(remove_html_tags)
```

```python
df['title']=df['title'].apply(remove_html_tags)
```

```python
#Remove panctuation
import string
exclude = string.punctuation
```

```python
def remove_punc(df):
    return df.translate(str.maketrans("","",exclude))
```

```python
df['title']=df['title'].apply(remove_punc)
```

```python
df['content']=df['content'].apply(remove_punc)
```

```python
df
```

```python
# short word treatment
```

```python
def text_conversation(df):
    new_df = []
    for w in df.split():
        if w.upper() in df:
            new_df.append(df[w.upper()])
        else:
            new_df.append(w)
    return " ".join(new_df)
```

```python

df['content']=df['content'].apply(text_conversation)
df['content']
```

```python
import re
def tokenization(text):
    tokens = re.split('W+',text)
    return tokens
#applying function to the column
df['content']= df['content'].apply(lambda x: tokenization(x))
```

```python
df['title']=df['title'].apply(lambda x: tokenization(x))
```

```python
df
```

```python
import nltk
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
#['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're"]
```

```python
#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
```

```python
#applying the function
df['title']= df['title'].apply(lambda x:remove_stopwords(x))
df['content']=df['content'].apply(lambda x:remove_stopwords(x))
df
```

```python
df=df.astype(str)
print(df.dtypes)
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
df['embeddings']=df['title'].apply(model.encode)
```

```python
df['Embeddings']=df['content'].apply(model.encode)
```

```python
df
```

```python
x=df['embeddings'].to_list()
y=df['source'].to_list()
```

```python
# Logistic Regression model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.1,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train, y_train)
```

```python
from sklearn import metrics 
predicted=LR.predict(x_test)
print('Logistic Regression Accuracy:', metrics.accuracy_score(y_test, predicted))

print('Logistic Regression Precision:', metrics.precision_score(y_test, predicted,pos_label='positive',average='macro'))
print('Logistic Regression Recall:', metrics.recall_score(y_test, predicted,pos_label='positive',average='micro'))

```
